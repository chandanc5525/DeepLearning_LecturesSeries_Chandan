## **Edge Detection Filter**


```python
# Sobel Filter:
'''
- Detects edges by calculating gradients in x and y directions.

- cv2.CV_64F:

Output image data type (64-bit float) to handle negative values 1, 0

- ksize=5: Uses a 5x5 kernel instead of default 3x3

'''

sobelx = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=5)
sobely = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=5)
sobelxy = cv2.Sobel(image, cv2.CV_64F, 1, 1, ksize=5)

plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.imshow(sobelx, cmap='gray')
plt.title('Sobel X')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(sobely, cmap='gray')
plt.title('Sobel Y')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(sobelxy, cmap='gray')
plt.title('Sobel XY')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Using Canny edges
edges = cv2.Canny(image, 100, 200)

# Find contours
contours, hierarchy = cv2.findContours(edges, cv2.RETR_EXTERNAL,
                                       cv2.CHAIN_APPROX_SIMPLE)

# Create a copy for drawing contours
if len(image.shape) == 2 or image.shape[2] == 1:
    # Image is grayscale
    img_contours = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)
else:
    # Image is already BGR
    img_contours = image.copy()

# Draw contours
cv2.drawContours(img_contours, contours, -1, (0,255,0), 1)

plt.imshow(cv2.cvtColor(img_contours, cv2.COLOR_BGR2RGB))  # Convert for matplotlib
plt.title("Contours")
plt.axis('off')
plt.show()
```

```python
# Using Canny edges
edges = cv2.Canny(image, 100, 200)

# Find contours
contours, hierarchy = cv2.findContours(edges, cv2.RETR_EXTERNAL,
                                       cv2.CHAIN_APPROX_SIMPLE)

# Create blank black canvas
img_contours = np.zeros((image.shape[0], image.shape[1], 3), dtype=np.uint8)

# Draw contours in green
cv2.drawContours(img_contours, contours, -1, (0,255,0), 1)

plt.imshow(img_contours)
plt.title("Contours")
plt.axis('off')
plt.show()
```

```python
img = cv2.imread('/content/Albert.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Load pre-trained face detector
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Detect faces
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

# Draw rectangles around faces
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Face Detection")
plt.axis('off')
plt.show()
```
